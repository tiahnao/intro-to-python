```python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data related to electricity consumption
challenge_data = {
    'Household_ID': np.arange(1, 201),
    'Electricity_Consumption_kWh': np.random.normal(200, 50, 200),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=200),
    'Appliance_Count': np.random.randint(1, 10, 200),
    'Monthly_Bill_USD': np.random.uniform(50, 300, 200),
    'Household_Members': np.random.randint(1, 6, 200),
    'Has_Solar_Panels': np.random.choice(['yes', 'no'], size=200)  # New column for solar panels
}

# Introduce missing values and outliers
challenge_data['Electricity_Consumption_kWh'][10] = np.nan
challenge_data['Electricity_Consumption_kWh'][80] = 400  # Introduce an outlier
challenge_data['Monthly_Bill_USD'][30:60] = np.nan  # Introduce additional missing values in a range
challenge_data['Monthly_Bill_USD'][150:170] = 500  # Introduce outliers in another range

# Create a DataFrame
challenge_df = pd.DataFrame(challenge_data)

challenge_df
```
